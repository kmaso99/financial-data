```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

LOADING THE DATASET

```python
df = pd.read_csv("C:/Users/Kate/Documents/BCGX_data_file.csv")
```

READING THE MANUALLY EXTRACTED DATA FROM THE SEC EDGAR DATABASE

https://www.sec.gov/edgar/searchedgar/cik

```python
df
```

CALCULATING YEAR-BY-YEAR GROWTH RATES FOR "TOTAL REVENUE" AND "NET INCOME"

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df
```

INPUTING NULL/NaN VALUES THAT RESULT FROM pct_change CALUCULATIONS WITH 0

```python
df.fillna(0, inplace = True)
```

```python
df
```

CALCULATING YEAR-BY-YEAR GROWTH RATES FOR "TOTAL ASSETS", "TOTAL LIABILITIES", AND "CASH FLOW FROM OPERATING ACTIVITES"

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow from Operations Growth(%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change()
```

```python
df
```

```python
df.fillna(0, inplace = True)
```

```python
df
```

```python
df.to_csv('final_data_report.csv')
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow from Operations Growth(%)' :'mean'    
}).reset_index()
```
